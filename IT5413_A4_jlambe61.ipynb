```python
# opening file and writing to the file

writing_file = open('golf.txt', 'w')

# importing linesep so program can be run across differnt operating systems without issue

from os import linesep

# setting up a loop to ask for input from user for the 16 players

for num in range(0, 16):
    
    first_name = input('Enter the players first name: ')
    last_name = input('Enter the players last name: ')
    handicap = input('Enter the players handicap: ')
    score = input('Enter the players score: ')
    
    # combining the user input into a string
    
    player_data = [first_name, last_name, handicap, score]
    data = ' '.join(player_data)

    # writing the string into the file and using linesep to put each string onto its own line
    writing_file.write(data + linesep)

# closing the file to save the input
    
writing_file.close()
```

```python
# opening the file and reading the data

reading_file = open('golf.txt', 'r')

# creating a header to display the data from the file
# I chose to combine first and last name into a single field

print('%-15s%-10s%-10s%-10s' % ('Name', 'Handicap', 'Score', 'Par'))

# using a for loop to read the lines of data from the file

for line in reading_file.readlines():
 
    # splitting the lines into individual data
    
    data = line.rstrip().split()

    # assigning each data point a variable based on its location in the string
    
    first_name = data[0]
    last_name = data[1]
    handicap = float(data[2])
    score = int(data[3])
    
    # this code will determine whether the player was at, above, or below par
    
    par = 80

    par_value = ''
    if score == par:
        par_value = 'Made Par'
    elif score < par:
        par_value = 'Under Par'
    else:
        par_value = 'Over Par'

    # populating the list with data
    # I combined the names into one heading
    # since handicap was a float and score was an int I had to apply a decimal value to my spacing
    
    print('%-15s%-10.1f%-10.0i%-10s' % (first_name + ' ' + last_name, handicap, score, par_value))

```
