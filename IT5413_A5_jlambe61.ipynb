```python
#defining the lists for rainfall amount and months of the year

rain_fall = []
months_of_year = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

#asking for input from the user for the rainfall list

for item in months_of_year:
    print('Please enter the amount of rainfall for',item,':')
#converting user input to float data and adding it to the rainfall list

    n = (input())
    n = float(n)
    
    rain_fall.append(n)
    
#creating a heading for my list

print('%-10s%-10s' % ('Month', 'Rainfall'))

#joining the lists together 

combined = list(zip(months_of_year, rain_fall))

for item in combined:
    print(item)
    
#calculating the sum of monthly rainfall for yearly rainfall    
    
sum_ = sum(rain_fall)

print('The total rainfall for the year was %.2f'%sum_, 'inches')

#calculating the sum for average monthly rainfall
    
avg_ = sum_/len(rain_fall)

print('The average rainfall each month was %.2f'%avg_, 'inches')

#using a for loop to identify the max and min items in the rainfall list and pairing it to the correct month

for i in range(12):
    if(rain_fall[i]==max(rain_fall)):
        print(months_of_year[i],'had the most rainfall with', rain_fall[i],'inches')
        
for i in range(12):
    if(rain_fall[i]==min(rain_fall)):
        print(months_of_year[i],'had the least rainfall with', rain_fall[i],'inches')
```
